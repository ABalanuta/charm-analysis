```python
# battery life curve generation

sensi_lorasim = pd.DataFrame(
        {7:{ 125:-126.5,  250:-124.25, 500:-120.75}, 
         8:{ 125:-127.25, 250:-126.75, 500:-124.0}, 
         9:{ 125:-131.25, 250:-128.25, 500:-127.5}, 
         10:{125:-132.75, 250:-130.25, 500:-128.75}, 
         11:{125:-134.5,  250:-132.75, 500:-128.75}, 
         12:{125:-135.25, 250:-133.25, 500:-132.25}}).transpose()

# log shadow model parameters
Ptx = 15 # in dBm
# gamma = 2.2394 # from campus test
gamma= 2.14949648 # from drive test
# d0 = 40.0 # in m
d0 = 40.0 # from drive tests
var = 0 # variance ignored for now. 
# Lpld0 = 127.41
# Lpld0 = 95.0038 # from campus test
Lpld0 = 98.0729471 # from drive test
estimatedL0 = 107.812891686 # from estimated correction to sensitivity values
GL = 0

msgFreq = 5 # per hour
AA_capacity = 3000*3600 # mA-sec
batteryLimitRatio = 0.9 # ratio of battery that can be used
daysPerYear = 365
TXPow = 121.00 # energy consumed during TX
node_power = -123.6 # original power of node
mean_gain = np.array([0, 1.5081, 1.6200, 2.5012, 2.8387, 3.0194, 3.1146, 3.1529])
nGateways = np.arange(1,8+1,1)
nodeGain = node_power + mean_gain


# the sensitivity values have to be corrected to fit our noise levels
L0_correction =  estimatedL0 - Lpld0
sensi_1301 = sensi_lorasim + L0_correction

def airtime(sf,cr,pl,bw):
    H = 0        # implicit header disabled (H=0) or not (H=1)
    DE = 0       # low data rate optimization enabled (=1) or not (=0)
    Npream = 8   # number of preamble symbol (12.25  from Utz paper)

    if bw == 125 and sf in [11, 12]:
        # low data rate optimization mandated for BW125 with SF11 and SF12
        DE = 1
    if sf == 6:
        # can only have implicit header with SF6
        H = 1

    Tsym = (2.0**sf)/bw
    Tpream = (Npream + 4.25)*Tsym
    payloadSymbNB = 8 + max(math.ceil((8.0*pl-4.0*sf+28+16-20*H)/(4.0*(sf-2*DE)))*(cr+4),0)
    Tpayload = payloadSymbNB * Tsym
    return Tpream + Tpayload

def toBatteryLife(sf, cr, pl, bw, msgFreq):
    msgTime = airtime(sf,cr,pl,bw)/1000
    msgPow = TXPow*msgTime
    powPerYear = msgPow*msgFreq*24*365
    life = AA_capacity*batteryLimitRatio/powPerYear
    return life
    
def getNodeBatteryLife(initPower, msgFreq):
    gain = initPower + mean_gain
    supportedSF = np.zeros(gain.shape)
    for sf in range(12, 6, -1):
        supportedSF[gain > sensi_1301.loc[sf,125]] = sf
    nodeLife = np.zeros(supportedSF.shape)
    for i, sf in enumerate(supportedSF):
        nodeLife[i] = toBatteryLife(sf, 1, 20, 125, msgFreq)
    return nodeLife

# life0 = getNodeBatteryLife(-123.6, 2)
life1 = getNodeBatteryLife(-123.6, 5)
life2 = getNodeBatteryLife(-123.6, 10)
life3 = getNodeBatteryLife(-123.6, 15)

plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.size"] = 20
    
plt.figure()
# plt.plot(nGateways, life0, 'o-')
plt.plot(nGateways, life1, 'o-', label="5")
plt.plot(nGateways, life2, 'o-', label="10")
plt.plot(nGateways, life3, 'o-', label="15")
legend=plt.legend(title="msg per hour", fontsize='x-small', bbox_to_anchor=(0.6,1.015), loc='upper left')
plt.setp(legend.get_title(),fontsize='x-small')
plt.xlabel("Number of gateways")
plt.ylabel("Battery life (years)")
plt.ylim([0,10])
plt.xticks(np.arange(1,9,1))
plt.grid()
plt.tight_layout()
plt.savefig('diversity_battery.pdf', format='pdf')

plt.show()
```
